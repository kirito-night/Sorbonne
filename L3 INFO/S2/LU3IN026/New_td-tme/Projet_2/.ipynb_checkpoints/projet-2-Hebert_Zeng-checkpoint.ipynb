# Non supervisé

## Problématique

Sous les critères du changement climatique et de l'exposition aux particules fines, quelles sont les sous-catégories d'aliments qui sont les plus impactantes à produire ?

## Analyse

### Choix des critères et formatage des données

Le choix du changement climatique et des particules fines se base sur la robustesse des critères : la documentation Agribalyse nous indique que ces 2 indicateurs font partis des 3 les plus satisfaisants. Ils sont donc intéressant à étudier. Le but est aussi de voir si ces 2 indicateurs possèdent une certaine connexion.

Récupérons les colonnes des fichiers pour choisir celles qui nous concerne.

```python
print(data_synthese.columns, "\n")
print(data_ingredient.columns, "\n")
print(data_etape.columns)
```

Nous allons donc choisir le nom, le sous groupe, le changement climatique et les particules.
Avant cela, trions un peu les données de base. En effet il existe un indicateur appelé DQR qui évalue la fiabilité des données.
Un DQR > 3 souligne des données peu fiables. Nous allons donc les retirer.

```python
data_synthese_fiable = data_synthese[data_synthese["DQR - Note de qualité de la donnée (1 excellente ; 5 très faible)"] < 3.0]
data_etape_fiable = data_etape[data_etape["DQR"] < 3.0]
```

```python
data_synthese_fiable.loc[0:10]
```

```python
data_synthese_fiable_2D = data_synthese_fiable[["Nom du Produit en Français", "Sous-groupe d'aliment","Changement climatique (kg CO2 eq/kg de produit)", "Particules (E-06 disease inc./kg de produit)"]]
index = [i for i in range(len(data_synthese_fiable_2D))]
data_synthese_fiable_2D.index = index
data_synthese_fiable_2D
```

Nous avons donc notre fichier synthèse fiable en 2D sur lequel nous allons travailler. Commencons par normaliser les valeurs.

```python
temp = clust.normalisation(data_synthese_fiable_2D.iloc[:,2:4])
data_synthese_fiable_2D_norm = pd.concat([data_synthese_fiable_2D.iloc[:,0:2], temp], axis=1)
data_synthese_fiable_2D_norm
```

```python
plt.scatter(data_synthese_fiable_2D_norm.iloc[:,2],data_synthese_fiable_2D_norm.iloc[:,3])
```

### Algorithme k-moyennes

Définissons une fonction d'affichage pour l'algorithme des k-moyennes.

```python
def affiche_resultat(Base,Centres,Affect):
    for k ,v in Affect.items() : 
        b  = Base.iloc[v]
        plt.scatter(Centres[:,0],Centres[:,1],color='r',marker='x')
        plt.scatter(b["Changement climatique (kg CO2 eq/kg de produit)"],b["Particules (E-06 disease inc./kg de produit)"])
```

Nous allons donc maintenant procéder à l'analyse par l'algorithme des k-moyennes. Pour cela il nous faut trouver un nombre de clusters optimal. Nous allons utiliser la méthode elbrow suivi d'une analyse de l'évolution de l'Indice de Dunn pour le déterminer.

```python
liste_variances = []
liste_Dunn = []
for i in range(1,10):
    les_centres, l_affectation = clust.kmoyennes(i, data_synthese_fiable_2D_norm.iloc[:, 2:4], 0.05, 100)
    variance = 0
    for j in range(len(les_centres)):
        for k in range(len(l_affectation[j])):
            variance += (data_synthese_fiable_2D_norm.iloc[(l_affectation[j])[k],2] - les_centres[j][0])**2 + (data_synthese_fiable_2D_norm.iloc[(l_affectation[j])[k],3] - les_centres[j][1])**2
    liste_variances.append(variance)
    liste_Dunn.append(clust.indexDunn(data_synthese_fiable_2D_norm.iloc[:,2:4],les_centres,l_affectation))
```

La méthode elbrow nous idique que le "coude" du graphique se situe sur 2.

```python
x = [i for i in range(1,10)]
plt.plot(x, liste_variances)
```

L'Indice de Dunn nous indique aussi que 2 est le chiffre que l'on recherche, n'avoir qu'un seul cluster étant inutile pour une analyse correcte.

```python
x = [i for i in range(1,10)]
plt.plot(x, liste_Dunn)
```

Nous allons donc lancer l'algorithme pour 2 clusters.

```python
les_centres, l_affectation = clust.kmoyennes(2, data_synthese_fiable_2D_norm.iloc[:, 2:4], 0.05, 100)
```

```python
clust.indexDunn(data_synthese_fiable_2D_norm.iloc[:,2:4],les_centres,l_affectation)
```

```python
affiche_resultat(data_synthese_fiable_2D_norm,les_centres,l_affectation)
```

Nous avos donc bien nos 2 clusters différents autour des 2 centres.

```python
print(les_centres)
```

### Analyse des résultats

Nous allons donc déterminer quels sous-groupes d'aliments appartiennent à quel cluster.

Déterminons les sous-groupes d'aliments du cluster 1 en bleu.

```python
groupe1 = set()
clust1 = l_affectation[0]
for i in clust1:
    groupe1.add(data_synthese_fiable_2D_norm.iloc[i, 1])
print(groupe1)
```

Tous ces sous-groupes font donc partis du cluster le moins impactant, tant au niveau du changement climatique que des particules fines. On peut donc considérer que ce sont les moins dangereux à produire sur ces critères.

Déterminons les sous-groupes d'aliments du cluster 2 en orange.

```python
groupe2 = set()
clust2 = l_affectation[1]
for i in clust2:
    groupe2.add(data_synthese_fiable_2D_norm.iloc[i, 1])
print(groupe2)
```

Tous ces sous-groupes font donc partis du cluster le plus impactant, tant au niveau du changement climatique que des particules fines. On peut donc considérer que ce sont les plus dangereux à produire sur ces critères.

# Supervisé

## Problématique

Peut-on prévoir de façon précise le groupe d'aliment à partir des indicateurs de changement climatique, de l'appauvrissement de la couche d'ozone et des particules fines ?

## Analyse

### Choix des critères et formatage des données

Le choix du changement climatique, de l'appauvrissement de la couche d'ozone et des particules fines se base sur la robustesse des critères : la documentation Agribalyse nous indique que ces 3 indicateurs sont les plus satisfaisants. Ils sont donc intéressant à étudier.

Récupérons les colonnes des fichiers pour choisir celles qui nous concerne.

```python
print(data_synthese.columns, "\n")
print(data_ingredient.columns, "\n")
print(data_etape.columns)
```

Nous allons donc choisir le changement climatique, l'appauvrissement de la couche d'ozone et les particules. Avant cela, trions un peu les données de base. En effet il existe un indicateur appelé DQR qui évalue la fiabilité des données. Un DQR > 3 souligne des données peu fiables. Nous allons donc les retirer. Par sécurité nous n'allons garder que les données avec un DQR < 2.5 pour augmenter la précision de nos classifiers

```python
data_synth_fiable = data_synthese[data_synthese["DQR - Note de qualité de la donnée (1 excellente ; 5 très faible)"] < 2.5]
index = [i for i in range(len(data_synth_fiable))]
data_synth_fiable.index = index
```

```python
data_synth_fiable.loc[0:10]
```

```python
cols = [
    "Changement climatique (kg CO2 eq/kg de produit)",
    "Particules (E-06 disease inc./kg de produit)",
    "Appauvrissement de la couche d'ozone (E-06 kg CVC11 eq/kg de produit)"
    ]
data_synth_fiable_3D = data_synth_fiable[cols]
data_synth_fiable_3D
```

### Classifier Arbre de décision numérique

Comme prévu nous décidons de prédire la colonne "Groupe d'aliments".

```python
desc= np.array(data_synth_fiable_3D)
label = np.array(data_synth_fiable["Groupe d'aliment"])
```

Lançons le train de l'arbre numérique.

```python
import graphviz as gv
arbre = cl.ClassifierArbreNumerique(len(cols), 0.0 , cols)
arbre.train(desc, label)
print(arbre.accuracy(desc, label))
```

Sauvegardons l'arbre dans le format png.

```python
graphe_arbre_1 = gv.Digraph(format='png')
arbre.affiche(graphe_arbre_1)
graphe_arbre_1.render("tree")
```

Calculons maintenant la moyenne des accuracy pour évaluer la prédiction de la colonne.

```python
index = np.random.permutation(len(desc)) # mélange des index
XtoyMelange = desc[index]
YtoyMelange = label[index]

learn_rate = 0.01
seuil = 0.001

n_fold = 10

cross_acc_test = []


for i in range(n_fold):
    Xapp,Yapp,Xtest,Ytest = ut.crossval(XtoyMelange, YtoyMelange, n_fold, i)
    #print("========== ITERATION : ",i," ==========")
    #print(" Xapp=", Xapp.T,"\n","Yapp=", Yapp,"\n","Xtest=",Xtest.T,"\n","Ytest=", Ytest)
    tree_cross = cl.ClassifierArbreNumerique (len(cols) , 0 , cols) 
    tree_cross.train(Xapp , Yapp)
    
    res_test = tree_cross.accuracy(Xtest,Ytest)
    cross_acc_test.append(res_test)
    #print(res_test)

print("Accuracy des tests a chaque iteration",cross_acc_test, "\n")
print("Moyenne des accuracy :" , np.asarray(cross_acc_test).mean())
```

### Analyse des résultats

Tout d'abord, en analysant l'image de l'arbre (qui est certes extrêmement grande dû au nombre important de données), on remarque que les groupes d'aliments les plus impactant sur l'environnement sont les mêmes que ceux déterminés par analyse non supervisée. On peut donc conclure que le résultat est cohérent.
En étudiant la moyenne des accuracy on remarque que la précision n'est pas optimale. Cela peut être dû au choix de la colonne à prédire, qui est sûrement trop générale pour une prédiction précise. Il est possible que choisir une colonne moins large comme le sous-groupe d'aliment par exemple améliore cette précision. Cela peut aussi venir du choix des colonnes qui servent à l'analyse. Même si ce sont les 3 indicateurs les plus précis, leur fiabilité n'est pas absolue. Utiliser plus de colonnes peut améliorer le résultat mais il faut faire attention car les autres indicateurs sont encore moins fiables.
